```python
# -------------------------------
# DOCUMENT CLUSTERING USING K-MEANS
# -------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import normalize

import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

# --------------------------------------------
# 1. TEXT PREPROCESSING FUNCTION
# --------------------------------------------
def clean_text(text):
    text = text.lower()                             # lowercase
    text = re.sub(r'[^a-z\s]', '', text)            # remove punctuation & numbers
    words = text.split()
    words = [w for w in words if w not in stop_words]   # remove stopwords
    words = [stemmer.stem(w) for w in words]            # stemming
    return " ".join(words)

# --------------------------------------------
# EXAMPLE DOCUMENTS (You can replace with your own)
# --------------------------------------------
documents = [
    "India won the cricket match yesterday.",
    "The stock market crashed due to inflation fears.",
    "Virat Kohli scored a century in the finals.",
    "New AI technology is transforming industries.",
    "Sensex and Nifty showed heavy volatility today.",
    "Machine Learning and Deep Learning are evolving fast.",
    "Bitcoin price dropped after regulatory issues.",
    "The government announced new economic reforms."
]

# Clean all documents
cleaned_docs = [clean_text(doc) for doc in documents]

# --------------------------------------------
# 2. TF-IDF VECTORIZATION
# --------------------------------------------
vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = vectorizer.fit_transform(cleaned_docs)

# Normalize vectors (optional but improves clustering)
tfidf_matrix = normalize(tfidf_matrix)

# --------------------------------------------
# 3. ELBOW METHOD (To find best K)
# --------------------------------------------
inertia_values = []
K_range = range(2, 8)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia_values.append(kmeans.inertia_)

plt.plot(K_range, inertia_values, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()

# --------------------------------------------
# 4. K-MEANS CLUSTERING (Choose K)
# --------------------------------------------
K = 3   # choose based on elbow
model = KMeans(n_clusters=K, random_state=42)
labels = model.fit_predict(tfidf_matrix)

# --------------------------------------------
# 5. SILHOUETTE SCORE
# --------------------------------------------
sil_score = silhouette_score(tfidf_matrix, labels)
print("Silhouette Score:", sil_score)

# --------------------------------------------
# 6. TOP TERMS OF EACH CLUSTER
# --------------------------------------------
terms = vectorizer.get_feature_names_out()

print("\nTop Terms Per Cluster:")
for i in range(K):
    centroid = model.cluster_centers_[i]
    top_term_ids = centroid.argsort()[-10:]
    top_terms = [terms[t] for t in top_term_ids]
    print(f"Cluster {i}: {top_terms}")

# --------------------------------------------
# 7. SHOW DOCUMENTS WITH THEIR CLUSTERS
# --------------------------------------------
print("\nDocument → Cluster Mapping")
for doc, label in zip(documents, labels):
    print(f"[Cluster {label}] {doc}")
```
