```python
'''1. Label Encoding (For Ordinal Categories)

Use when categories have an order
Example: Low < Medium < High'''
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
# Sample dataset
data = pd.DataFrame({
    'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small']
})
# Initialize encoder
le = LabelEncoder()
# Encode
data['Size_Encoded'] = le.fit_transform(data['Size'])
print(data)
```

```python
'''2. One-Hot Encoding (For Nominal Categories)

Use when categories have no order
Example: Red, Blue, Green'''
```

```python
data = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Blue']
})

encoded_data = pd.get_dummies(data, columns=['Color'])

print(encoded_data)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Blue']
})

# Updated parameter name
encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(data[['Color']])

print(encoded)
```

```python
'''3. Ordinal Encoding (Manual Order)

If categories have meaningful ranking:'''
```

```python
from sklearn.preprocessing import OrdinalEncoder

data = pd.DataFrame({
    'Education': ['High School', 'Bachelor', 'Master', 'PhD']
})

encoder = OrdinalEncoder(
    categories=[['High School', 'Bachelor', 'Master', 'PhD']]
)

data['Education_Encoded'] = encoder.fit_transform(data[['Education']])

print(data)
```
